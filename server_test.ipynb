```python
import requests
from IPython.display import Audio


def fecth_and_play_voice(params):
    url = "http://127.0.0.1:5000/voice"
    response = requests.get(url, params=params)

    if response.status_code == 200:
        display(Audio(response.content))
    else:
        print(f"Error: {response.status_code}")
        print(response.json())


params = {
    "text": "",
    "model_id": 10,
    "style": "Neutral",
    "style_weight": 100
}

fecth_and_play_voice(params)
```
